```python
from wordcloud import WordCloud, STOPWORDS
import pandas as pd
from matplotlib import pyplot as plt
import requests
from bs4 import BeautifulSoup as bs
```

```python
url = "https://mjverse.in/2020/03/07/taking-a-look-at-the-past-and-the-future-of-mechanical-engineering//"
response = requests.get(url)
response.raise_for_status()
soup = bs(response.text, 'html.parser')
text = ''
for tag in soup.find_all("div", class_="entry-content"):
    soup2 = bs(str(tag), 'html.parser')
    for para in soup2.find_all('p'):
        text += para.text
```

```python
wordcloud = WordCloud(max_font_size=55, max_words=100, background_color='black').generate(text)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
